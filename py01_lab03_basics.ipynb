```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# %run D:/test.py
```

```python
# Column oriented to Data Frame
sales = [('account', ['Jones LLC', 'Alpha Co', 'Blue Inc']),
         ('Jan', [150, 200, 50]),
         ('Feb', [200, 210, 90]),
         ('Mar', [140, 215, 95]),
         ]
df = pd.DataFrame.from_items(sales)

print ('\n converted sales df')
print (df)

# Row  oriented to Data Frame
sales = [('Jones LLC',150, 200, 50 ),
         ('Alpha go',200, 210, 90) ,
         ('Blue Inc', 140, 215, 95),
              ]
labels = ['Account','Feb','Mar','Apr']
df = pd.DataFrame.from_records(sales)

print ('\n converted sales df')
print (df)
```

```python
# Row oriented Dictionary to data frame

sales = [{'account':'Jones LLC', 'Jan':150,'Feb':200,'Mar':140},
         {'account':'Alpha Co', 'Jan':200,'Feb':210,'Mar':215},
         {'account':'Blue Inc', 'Jan':50,'Feb':90,'Mar':95}]

df = pd.DataFrame(sales)

print ('\n converted sales df')
print (df)

# Column  oriented Dictionary to data frame
sales = {'account':['Jones LLC', 'Alpha Co', 'Blue Inc'],
        'Jan':[150, 200, 50],'Feb':[200, 210, 90],'Mar':[140, 215, 95]}      

df = pd.DataFrame.from_dict(sales)

print ('\n converted sales df')
print (df)
```
