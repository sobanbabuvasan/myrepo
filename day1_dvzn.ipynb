```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_mktg = pd.read_csv("D:/dat1/retailMarketing2.csv",sep = ',')
print df_mktg.shape
```

```python
df_mktg.head()
```

```python
# the inline prints the plot instead of showing the plot , to show the plot you can specify plt.show()
%matplotlib inline
plt.plot(df_mktg.AmountSpent,df_mktg.Salary)
plt.xlabel('Salary')
plt.ylabel('Amount Spent')
plt.title('Amount Spent by different Salary bucket')
# plt.show() 
```

```python
plt.subplot(1,2,1)
plt.plot(df_mktg.Salary, df_mktg.AmountSpent,'g')
plt.subplot(1,2,2)
plt.plot( df_mktg.AmountSpent,df_mktg.Salary,'m')

```

```python
print df_mktg['Salary'].describe()
```

```python
sns.kdeplot(df_mktg['Salary'],shade=True)
sns.plt.show()
sns.rugplot(df_mktg['Salary'])
sns.plt.show()
sns.distplot(df_mktg['Salary'],kde=False,rug=False)
sns.plt.show()
```

```python
sns.distplot(df_mktg['Salary'],kde=True,rug=True)
sns.plt.show()
```

```python
# Using lmplot to analyze accomodation status and Age features 
# towards Amount Spent compared to Salary
sns.lmplot(x='Salary',y='AmountSpent',hue='OwnHome',col='Age',markers=["o","x"],
             data=df_mktg)
sns.plt.show()
```

```python
df_mktg['AmountSpent'].describe()
```

```python
# Using lmplot to analyze marital status and gender features towards
# Amount Spent compared to Salary
sns.lmplot(x='Salary',y='AmountSpent',hue='Married',col='Gender',
           markers=["o","x"],data=df_mktg)
sns.plt.show()
```

```python
# Analyzing the trend based number of Children
sns.lmplot(x='Salary',y='AmountSpent',col='Children',data=df_mktg)
sns.plt.show()
```

```python
# Based on distance from the store , analyzing the history of customer
sns.lmplot(x='Salary',y='AmountSpent',row='History',col='Location',
           data=df_mktg)
sns.plt.show()    
```
