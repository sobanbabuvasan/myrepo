```python
import numpy as np
import pandas as pd
from ggplot import *
```

```python
df_store = pd.read_csv("D:/dat1/Salesstore.csv",sep = '\t')
print df_store.shape
```

```python
df_store.head()
```

```python
plot = ggplot(df_store,aes(x='Profit',y='Sales', label="Profit"))
print plot+ geom_point()+geom_text(check_overlap=True,size=10) + xlab("Profit")+ylab("Sales")+ylim(20,160)
```

```python
df_agg = df_store.groupby(['Ship_Mode'])['Sales'].sum()
print df_agg
```

```python
plot = ggplot(df_store,aes(x='Profit',y='Sales', label="Ship_Mode"))
print plot+ geom_point()+geom_text(check_overlap=True,size=10) + xlab("Profit")+ylab("Sales")+ylim(20,160)
```

```python
plot_bar = ggplot(df_agg,aes(x='Ship_Mode', weight='Sales'))
print plot_bar + geom_bar(stat='Identity') + xlab("Ship_Mode")
```
