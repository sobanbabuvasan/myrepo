```python
print df_mktg.groupby(['Age'])['Salary'].mean()
sns.barplot(x='Age',y='Salary',data=df_mktg)
sns.plt.show()

print df_mktg.groupby(['Age'])['AmountSpent'].mean()
sns.barplot(x='Age',y='AmountSpent',data=df_mktg)
sns.plt.show()
```

```python
    #Analyzing amount spent based on Gender and marital status
    #Note when giving orientation ,ensure the categorical variable and continuos variables are assigned on correct axis
    print df_mktg.groupby(['Gender','Married'])['AmountSpent'].mean()
    sns.barplot(x='AmountSpent',y='Gender',hue='Married',orient='h',data=df_mktg)
    sns.plt.show()
```

```python
#Box plot to analyze salary component of the customers based on the Marital status and the
#accomodation they are living i.e. Rent or Own
sns.boxplot(x='OwnHome',y='Salary',hue='Married',data=df_mktg,palette='Set3')
sns.plt.show()
```

```python
# Amount spent based on the Catalogs shared with the customers
sns.boxplot(x='Catalogs',y='AmountSpent',data=df_mktg)
sns.plt.show()
```

```python
# Individually over here , we  are also computing mean salary
# across various age groups using the group by function on the 
# data frame marketing .
# this way we will be able to correlate the number given by 
# the group by function and the similar data being plotted 
# on the barplot chart as well
print df_mktg.groupby(['Age'])['Salary'].mean()
sns.barplot(x='Age',y='Salary', data=df_mktg)
sns.plt.show()
# the bar shows the confidence interval
```
