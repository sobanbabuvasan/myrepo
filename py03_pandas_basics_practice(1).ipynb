#### How to drop columns and how to use applymap() and pipe()

```python
def hg(s):
    res = s + 10
    return res

df = ex10.drop(['devices1','response','bu','devices2','devices'], axis=1)
# print df

# applying the in-built function
df.applymap(np.sqrt)

# applying the lambda function
print '\n' 
print 'Lambda '
print df.applymap(lambda x:x*2)
print df.applymap(lambda x:x/2)
print '\n'
# apply the custom function hg over every cell in the dataframe
df.applymap(hg)

# Now apply hg function only on clicks and leads 
print df[['clicks','leads']]
print df[['clicks','leads']].applymap(hg)

# Now apply hg function only on 0,1,3 rows of clicks and leads columns
print df.loc[(0,1,3),['clicks','leads']].applymap(hg)

#print ex10.loc[0:3, lambda ex10: ['leads', 'clicks']]


def arith(x):
    # that, if x is a string,
    if type(x) is str:
        # just returns it untouched
        return x
    # but, if not, return it adding by  20
    elif x:
        return 20 +  x
    # and leave everything else
    else:
        return

print df.loc[(0,1,3),['clicks','leads']].applymap(arith)
print df.loc[(0,1,3),:].applymap(arith)

def arith2(a1,a2):
       return a1+a2
    
print df.pipe(arith2,2)
print df.pipe(hg)
```

### How to handle exception

```python
x = 10
y = 0
try:
    res = x/y
except Exception as e:
    print ('Exception occurred ', e)    
    res = None
print ('Result of Division is ', res)
```

```python
x = 10
y = 0
try:
    res = x/y
except ZeroDivisionError as e:
    print ('Exception occurred ', e)    
    res = None
print ('Result of Division is ', res)
```

```python
x = '10'

y = 2

try:

    res = x/y

except Exception as e:

    print ('Exception Type ', type(e).__name__)    

    res = None

print ('Result of Division is ', res)

('Exception Type ', 'TypeError')
('Result of Division is ', None)
```
