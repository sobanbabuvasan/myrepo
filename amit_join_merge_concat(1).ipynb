```python
df1 = {'name': ['lexus','ferrari','mercedez'],
        'color':['red','white','blue'],
        'year':['2012','2014','2015'] }
print (df1)
```

```python
df1_cars = pd.DataFrame({'name': ['lexus','ferrari','mercedez'],
        'color':['red','white','black'],
        'year':['2012','2014','2015'] },index = [0,1,2])
 # 
print (df1_cars)     
type(df1_cars)
```

```python
df2_cars = pd.DataFrame({'name': ['lexus','bmw','mercedez'],
        'color':['red','white','blue'],
        'year':['2013','2015','2016'] } ,index = [0,1,3] )

print (df2_cars)     
type(df2_cars)
```

```python
# concatenate the data frames row wise 
concat_df = pd.concat([df1_cars,df2_cars],axis=0)
print (concat_df)
```

```python
# concatenate the data frames column (axis=1) wise 
concat_df = pd.concat([df1_cars,df2_cars],axis=1)
print (concat_df)
```

```python
df_merge = pd.merge(df1_cars,df2_cars,how='inner',left_on='color',right_on='color')
print (df_merge)
```

```python
df_merge = pd.merge(df1_cars,df2_cars,how='right',left_on='color',right_on='color')
print (df_merge)
df_merge = pd.merge(df1_cars,df2_cars,on='color')
print (df_merge)
```

```python
df_merge = pd.merge(df1_cars,df2_cars,how='outer')
print (df_merge)
```

```python
df_join = df2_cars.join(df1_cars,how='inner',lsuffix='_left',rsuffix='_right')
    
print (df_join)
```
