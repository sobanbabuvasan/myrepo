```python
# Showcasing sales across various Regions
fn = sns.FacetGrid(df_store,col='Region')
fn = fn.map(plt.hist,"Sales")
sns.plt.show(fn)    
```

```python
# Showcasing sales across all the shipping modes and the Product Containers used
fn = sns.FacetGrid(df_store,col='Ship_Mode',row='Product_Container',palette="GnBu_d",gridspec_kws={"width_ratios": [4, 4]}, size=2.5,aspect=3)
fn = fn.map(sns.distplot,"Sales",hist=False,rug=True)
sns.plt.show(fn)
```

```python
#Plotting bi-variate function on each facet ,keeping Profit and Sales as the baseline and analyzing the pattern shown by
# different Product Category across various Customer Segments

fn = sns.FacetGrid(df_store,col='Customer_Segment',hue='Product_Category',hue_kws=dict(marker=["^", "v","*"]),size=5)
fn = (fn.map(plt.scatter,"Profit","Sales").add_legend())

sns.plt.show(fn)

```
