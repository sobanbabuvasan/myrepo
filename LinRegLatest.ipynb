```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("D:/dat1/HousingS.csv")
```

```python
df.describe()
```

```python
sns.pairplot(df,x_vars = ['size','uds','park'],y_vars = ['price'],kind='reg',aspect=0.7,size=7)
```

```python
inputFeat = ['size','uds','park']
x = df[inputFeat]
y = df['price']
from  sklearn.cross_validation import train_test_split
x_train, x_test,y_train,y_test =  train_test_split(x,y,test_size=.30,random_state=1)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
ols1 = LinearRegression()
fit = ols1.fit(x_train,y_train)
```

```python
pred = fit.predict(x_test)
```

```python
pred
```

```python
from sklearn import metrics
print (metrics.mean_squared_error(y_test,pred) )
print (metrics.mean_absolute_error(y_test,pred) )
```

```python
print np.sqrt(metrics.mean_squared_error(y_test,pred))
```

```python
def compute_rsq(act,pred):
    tst = np.square(act - np.mean(act)).sum()
    rsq = 1.0 - ( np.square(act - pred).sum() / tst)
    return rsq
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,pred)
```
